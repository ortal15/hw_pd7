```python
import numpy as np
from IPython.display import display
import pandas as pd
```

```python
df = pd.read_csv("Sales_Funnel_CRM.csv")
df
```

```python
#● Show in a new Dataframe how many licenses sold for each company from each product
pivot = pd.pivot_table(df,values = 'Company' , index = 'Licenses' , columns ='Product' , aggfunc=lambda x: x.nunique())
pivot
```

```python
#● Show in a new Dataframe for each company what is the total amount of sale price and licences sold
pivot_2 = pd.pivot_table(df,values = 'Sale Price' , index = 'Company' , columns ='Licenses' , aggfunc='sum')
pivot_2
```
